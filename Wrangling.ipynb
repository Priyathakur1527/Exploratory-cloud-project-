Remove the duplicate values:

```python
df = df.drop_duplicates()
```

Normalize and standardize data

```python
# Normalize Total_Purchase_Amount and Frequency_of_Purchase
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df[["Total_Purchase_Amount", "Frequency_of_Purchase"]] = scaler.fit_transform(
    df[["Total_Purchase_Amount", "Frequency_of_Purchase"]]
)

# View normalizedÂ data
df.head()
```

```python
# Create a new feature for Customer Value: Total_Purchase_Amount * Frequency_of_Purchase
df["Customer_Value"] = df["Total_Purchase_Amount"] * df["Frequency_of_Purchase"]

# Calculate customer tenure in years
df["Customer_Tenure_Years"] = df["Customer_Tenure"] / 12

# View new features
df[["Customer_Value", "Customer_Tenure_Years"]].head()
```

```python
# Validate the final dataset
print("Final Dataset Info:")
df.info()

# Save the cleaned dataset
df.to_csv("cleaned_wrangling_dataset.csv", index=False)
print("Cleaned dataset savedÂ successfully!")
```
